From the picture of the given network we can extract its parameters:

$$
\begin{equation}
\textbf{W}_1 = \begin{bmatrix}
-2 & -1 \\
1 & 3
\end{bmatrix}, \,\,\, \textbf{B}_1 = \begin{bmatrix}
-0.5 \\
-0.5
\end{bmatrix} \\
\end{equation}
$$

$$
\begin{equation}
\textbf{W}_2 = \begin{bmatrix}
1& 1
\end{bmatrix}, \,\,\, b_2 = 0.5
\end{equation}
$$

Since we are using _purelin_ as activation functions we'll have $a^m_i = n^m_i$, where $m$ is the index layer, and $i$ is the neuron's index in the layer. The activations in the first layer will then be:

$$
\begin{align}
a^1_1 &= n^1_1 = -2p_1 - p_2 - 0.5 \\
a^1_2 &= n^1_2 = p_1 + 3p_2 - 0.5
\end{align}
$$

While in the second layer:

$$
\begin{align}
a^2_1 &= n^2_1 = a^1_1 + a^1_2 + 0.5 \\
\end{align}
$$

Replacing the values for $a^1_1$ and $a^1_2$:

$$
\begin{align}
a^2_1 &= n^2_1 = \left(-2p_1 - p_2 - 0.5\right) + \left(p_1 + 3p_2 - 0.5\right) + 0.5 \\ \\
a^2_1 &= n^2_1 = -p_1 + 2p_2 - 0.5 \\
\end{align}
$$

Therefore, the given 2-layer neural network will produce the same output as the 1-layer network with the following parameters:

$$
\begin{equation}
\textbf{W} = \begin{bmatrix}
-1 & 2
\end{bmatrix} , \,\,\, b = -0.5
\end{equation}
$$
